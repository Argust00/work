```python
import random
```

```python
def price(rm,k):
    return k/rm
```

```python
X_rm = X[:, 12]
k = random.randint(-500, 500)

price_by_random_k = [price(r, k) for r in X_rm]

draw_rm_and_price()
plt.scatter(X_rm, price_by_random_k)
```

```python

def loss(y,y_hat):
    return sum(abs(y_i - y_hat_i) for y_i, y_hat_i in zip(list(y), list(y_hat))) / len(list(y))
```

```python
trying_times = 20000

min_loss = float('inf')
best_k, best_b = None, None

for i in range(trying_times):
    k = random.random() * 200 - 100
    
    price_by_random_k = [price(r, k) for r in X_rm]

    current_loss = loss(y, price_by_random_k)
    
    if current_loss < min_loss:
        min_loss = current_loss
        best_k, best_b = k, b
        print('When time is : {}, get best_k: {}, and the loss is: {}'.format(i, best_k,min_loss))
```

```python
trying_times = 20000

min_loss = float('inf')

best_k = random.random() * 200 - 100


direction = [
    (+1),  # first element: k's change direction, second element: b's change direction
    
    (-1)
]

next_direction = random.choice(direction)

scalar = 0.1

for i in range(trying_times):
    
    k_direction = next_direction
    
    current_k = best_k + k_direction * scalar
    
    price_by_k = [price(r, current_k) for r in X_rm]

    current_loss = loss(y, price_by_k)
    
    if current_loss < min_loss: # performance became better
        min_loss = current_loss
        best_k = current_k
        
        next_direction = next_direction
        print('When time is : {}, get best_k: {}, and the loss is: {}'.format(i, best_k, min_loss))
    else:
        next_direction = random.choice(direction)
```
